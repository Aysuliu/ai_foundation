```python
data_with_missing = {
    "cars": ["BMW", "Volvo", None, "Tesla", "Ferrari", None, "Ford"],
    "counts": [3, 7, None, 8, 1, 9, None],
}

df_missing = pd.DataFrame(data_with_missing)
df_missing
```

```python
df_missing.isnull().sum() #boolean check
```

```python
df_missing.dropna(how="all") #drop missing values
df_missing.dropna(axis=1)
df_missing.dropna(subset=["cars"])
```

```python
df_missing.fillna({"cars":"unknown", "counts":0})
```
