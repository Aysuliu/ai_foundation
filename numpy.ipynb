```python
import numpy as np
arr = np.array([1,3,5,7])
print('numpy_array:', arr)

arr.shape
```

```python
arr2=np.array([[1,3,5,7],[4,6,8,9]])
print('numpy_array2:', arr2)
arr2.shape
```

```python
#dtype()
a = np.array([1,2,3])
print(a.dtype)
```

```python
#reshape()
a = np.array([1,2,3,4,5,6])
b = a.reshape(2,3)
print(b)
```

```python
#astype()
a = np.array([1,2,3])
b = a.astype(float)
print(b)
```

```python
#np.zeros(), np.ones() => dummy data
a = np.zeros((2,3))
b = np.ones((2,3))
print(a)
print(b)
```

```python
#ndim
a = np.array([1,2,3])
b = np.array([[1,2,3],[4,5,6]])
print(a.ndim)
print(b.ndim)
```

```python
#arange()
a = np.arange(1,6)
print(a)
```

```python
#sum(), mean(), max(), min()
a = np.array([1,2,3,4,5])
print(a.sum())
print(a.mean())
print(a.max())
print(a.min())
```

```python
#flatten()
a = np.array([[1,2],[3,4]])
print(a.flatten())
```

```python
# T ( Transpose)
a = np.array([[1,2],[3,4]])
print(a.T)
```

```python
# dot() @
a = np.array([[1,2],[3,4]])
b = np.array([[2,0],[1,2]])
print(a.dot(b))
#print(a@b)
```

```python
#random.rand() => 0.0 and 1.0 / 0.0 included but 1 excluded
a = np.random.rand(2,3)
print(a)
```
